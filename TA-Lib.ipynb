```python
import numpy as np
import copy
import yfinance as yf
import talib
```

```python
tickers = ["AMZN"]
ohlc_data = {}

for ticker in tickers:
    data = yf.download(ticker, period = '1y', interval="1d")
    data.dropna(how='any', inplace=True)
    data.columns = data.columns.droplevel(1)
    ohlc_data[ticker] = data
```

```python
dp = {}
for ticker in ohlc_data:
    df = ohlc_data[ticker].copy()
    df["3INSIDE"] = talib.CDL3INSIDE(df["Open"],df["High"],df["Low"],df["Close"])
    dp[ticker] = df
```

```python
signals = {}
for ticker in dp:
    df = dp[ticker]
    signals[ticker] = df[df["3INSIDE"] != 0 ]
```
