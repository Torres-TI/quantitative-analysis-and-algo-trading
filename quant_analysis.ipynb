```python
import datetime as dt
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stocks = ["AMZN", "MSFT", "META", "GOOG", "TSLA", "AAPL", "NVDA"]
start = dt.datetime.today() - dt.timedelta(3650)
end = dt.datetime.today()
cl_price = pd.DataFrame()
ohlcv_data = {}
```

```python
for ticker in stocks:
    cl_price[ticker] = yf.download(ticker,start,end)["Close"]
```

```python
cl_price.dropna(axis=0, inplace=True)
```

```python
daily_return = cl_price.pct_change() #Calculate the price change return actual_value / previous_value  - 1 
daily_return
```

```python
price_change_return = cl_price/cl_price.shift(1) - 1  #Calculate Price Change Return in Percentage
price_change_return
```

```python
df = daily_return.rolling(window=10).mean()
df.dropna()
```

```python
daily_return.ewm(com=10, min_periods = 10).mean() #exponential moving averages
```

```python
cumulative_return = (1+daily_return).cumprod() #cumprod, cumsum is a cumulative function
```

```python
simulation = 1000 * cumulative_return
```
