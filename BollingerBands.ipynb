```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ["AMZN"]
ohlcv_data = {}
```

```python
for ticker in tickers:
    data = yf.download(ticker, interval="5m", period="1D")
    data.columns = data.columns.droplevel(1)
    data.dropna(how="any", inplace=True)
    ohlcv_data[ticker] = data 
    
```

```python
def Boll_Band(DF, n=20):
    df = DF.copy()
    df["MB"] = df["Close"].rolling(n).mean()
    df["UB"] = df["MB"] + (df["Close"].rolling(n).std()*2)
    df["BB"] = df["MB"] - (df["Close"].rolling(n).std()*2)
    return df[["MB","UB","BB"]]

for ticker in ohlcv_data:
    ohlcv_data[ticker] = Boll_Band(ohlcv_data[ticker])
       
```
